```python
#Q1 Load dataset and create an ensemble ML model for predicting target variable (Result). Report the performance of the model using appropriate metrics.
import pandas as pd
local_path = r'E:\2-学习\1-Deakin\24-T1\SIT720\Task\data_9.csv'
df=pd.read_csv(local_path, comment='#')
df
```

```python
from scipy.stats import zscore
# Columns to z-score, excluding 'Result'
columns_to_zscore = df.columns.difference(['Result'])

# Z-score the selected columns
df[columns_to_zscore] = df[columns_to_zscore].apply(zscore)

# Display the DataFrame
df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score

# Split the data into a training set and a test set
X = df.drop('Result', axis=1)
y = df['Result']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Create a RandomForestClassifier
clf = RandomForestClassifier(n_estimators=100, random_state=42)

# Train the classifier
clf.fit(X_train, y_train)

# Make predictions on the test set
y_pred = clf.predict(X_test)

# Report the performance of the classifier
print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
#Have you used any hyperparameter tuning while building the model in Q1? If yes, then plot the performances that were obtained at different steps of optimisation. Otherwise, create an optimised model and compare performance with Q1.
from sklearn.model_selection import GridSearchCV

# Define the parameter grid
param_grid = {
    'n_estimators': [200, 100, 50],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10]
}

# Create a RandomForestClassifier
clf = RandomForestClassifier(random_state=42)

# Create a GridSearchCV object
grid_search = GridSearchCV(estimator=clf, param_grid=param_grid, cv=3, scoring='accuracy')

# Fit the GridSearchCV object to the data
grid_search.fit(X_train, y_train)

# Get the best parameters
best_params = grid_search.best_params_

# Create a RandomForestClassifier with the best parameters
clf_best = RandomForestClassifier(**best_params, random_state=42)

# Train the classifier
clf_best.fit(X_train, y_train)

# Make predictions on the test set
y_pred_best = clf_best.predict(X_test)

# Report the performance of the classifier
print("Accuracy (optimized):", accuracy_score(y_test, y_pred_best))
print(classification_report(y_test, y_pred_best))

# Compare with the previous model
print("Accuracy (previous):", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Convert the cv_results_ to a DataFrame
cv_results_df = pd.DataFrame(grid_search.cv_results_)

# Extract the columns with the mean test scores
mean_test_scores = cv_results_df[['param_n_estimators', 'param_max_depth', 'param_min_samples_split', 'mean_test_score']]

# Pivot the DataFrame to get the mean test scores for each combination of parameters
pivot_df = mean_test_scores.pivot_table(index=['param_n_estimators', 'param_min_samples_split'], columns='param_max_depth', values='mean_test_score')

# Plot the mean test scores
plt.figure(figsize=(10, 8))
sns.heatmap(pivot_df, annot=True, cmap='YlGnBu')
plt.title('Mean Test Scores for Different Parameter Combinations')
plt.show()
```

Q3 Reflect on the importance of hyperparameter tuning of ML models based on your ML model development exercise.  
Hyperparameter tuning is a critical step in the development of machine learning models. It can significantly improve the performance of a model by optimizing the parameters that define the model structure or how the model is trained.

In the provided code, GridSearchCV is used to perform hyperparameter tuning on a RandomForestClassifier. The parameters tuned include 'n_estimators', 'max_depth', and 'min_samples_split'. These parameters control the number of trees in the forest, the maximum depth of the trees, and the minimum number of samples required to split an internal node, respectively.

According to the optimization process graph, we can find the best hyperparameter to get larger score. By tuning these parameters, we can find the optimal balance between bias and variance, preventing overfitting or underfitting, and ultimately improving the model's predictive performance. For example, increasing 'n_estimators' can reduce the variance of the model, but it also increases the computational cost. Similarly, adjusting 'max_depth' can help control the complexity of the model and prevent overfitting.

In the exercise, the optimized model's performance was compared with the initial model, demonstrating the effectiveness of hyperparameter tuning. However, it's important to note that hyperparameter tuning should be performed carefully to avoid overfitting the validation set, as the best parameters found may not generalize well to unseen data.
