```python
# Q2 Select the optimum k value for clustering using Silhouette Coefficient and plot the optimum k values
cols = ["feature_" + str(i + 1) for i in range(df.shape[1] - 1)]
cols.insert(1, "Target")  # Insert "Target" as the second column name
df.columns = cols

df.head()
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

X=df.iloc[:,0:-1]
y=df.iloc[:,1]

silhouette_scores = []

# Determine the range of K values to explore
k_values = range(2, 20)

# Fit K-means models and calculate Silhouette Coefficient for each K value
for k in k_values:
    kmeans = KMeans(n_clusters=k, n_init=10)
    kmeans.fit(X)
    silhouette_avg = silhouette_score(X, kmeans.labels_)
    silhouette_scores.append(silhouette_avg)

# Find the optimal K value that maximizes the Silhouette Coefficient
optimal_k = k_values[silhouette_scores.index(max(silhouette_scores))]
print(f'Optimal K value: {optimal_k}')

# Plot the Silhouette Coefficient for each K value
plt.plot(k_values, silhouette_scores, marker='o')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Silhouette Coefficient')
plt.title('Silhouette Coefficient for Different Values of K')
plt.show()
```

According to the above graph, we can see the highest silhouette score is k=2.  

```python
# Q3 Create clusters using Kmeans and Kmeans++ algorithms with the optimal k value found in the previous problem. Report performances using appropriate evaluation metrics. Compare the results.

from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, pairwise_distances_argmin_min
from sklearn.metrics.cluster import contingency_matrix
import numpy as np

# Fit K-means and K-means++ models with the optimal K value
optimal_k =  2 # According to the Q2 result
kmeans = KMeans(n_clusters=optimal_k, n_init=10)
kmeans_pp = KMeans(n_clusters=optimal_k, init='k-means++', n_init=10)

# Fit models and predict cluster labels
kmeans.fit(X)
kmeans_pp.fit(X)
labels_kmeans = kmeans.labels_
labels_kmeans_pp = kmeans_pp.labels_

# Evaluate clustering performance using silhouette_score
silhouette_score_kmeans = silhouette_score(X, labels_kmeans)
silhouette_score_kmeans_pp = silhouette_score(X, labels_kmeans_pp)

# Evaluate clustering performance using purity
def calculate_purity(labels, y):
    contingency = contingency_matrix(y, labels)
    return np.sum(np.amax(contingency, axis=0)) / np.sum(contingency)

purity_kmeans = calculate_purity(labels_kmeans, y)
purity_kmeans_pp = calculate_purity(labels_kmeans_pp, y)

# Compare the results
print("Evaluation Metrics for K-means:")
print(f"Silhouette Score: {silhouette_score_kmeans}")
print(f"Purity: {purity_kmeans}")

print("\nEvaluation Metrics for K-means++:")
print(f"Silhouette Score: {silhouette_score_kmeans_pp}")
print(f"Purity: {purity_kmeans_pp}")
```

According to the result, when k=2, the Silhouette Score and Purity of K-means and K-means++ are same. There may be several reasons: 1. Data Characteristics, the dataset may have a simple structure, causing both K-means and K-means++ to produce similar clustering results. In such cases, the initialization strategy might not significantly impact the outcome. 2. K-means algorithms utilize random initialization for cluster centroids. Hence, different runs with random initializations may lead to similar clustering results in terms of evaluation metrics. 3. Parameter Settings, Adjusting parameters like the number of clusters (K), the number of iterations, or initialization strategies might reveal differences in clustering results between K-means and K-means++. 4. Clustering Evaluation Methods, Silhouette Score and Purity are two distinct clustering evaluation metrics that measure clustering quality from different perspectives. In some scenarios, both metrics might yield similar results, particularly when the dataset's structure is straightforward.
If change the optimal k, we can see Silhouette Score and Purity changes, and K-means's result is not always same as K-means++.

```python
# Q5 Apply DBSCAN on this dataset ("Live_20210128.csv") and find the optimum "eps" and "min_samples" value. Is the number of cluster same as the cluster found in Q2? Explain the similarity or differences that you have found between two solutions. 
from sklearn.cluster import DBSCAN
from sklearn.metrics import silhouette_score
from sklearn.metrics.cluster import contingency_matrix
import numpy as np

# Define ranges for eps and min_samples
eps_range = range(2, 20)
min_samples_range = range(2, 20)

best_eps = None
best_min_samples = None
best_silhouette_score = -1
best_purity = -1

# Iterate over eps and min_samples values
for eps in eps_range:
    for min_samples in min_samples_range:
        # Perform DBSCAN clustering
        dbscan = DBSCAN(eps=eps, min_samples=min_samples)
        labels = dbscan.fit_predict(df)
        
        # Check if more than one cluster is formed
        if len(set(labels)) > 1:
            # Compute silhouette score
            silhouette_avg = silhouette_score(df, labels)
            
            # Compute purity
            contingency = contingency_matrix(y, labels)
            purity = np.sum(np.amax(contingency, axis=0)) / np.sum(contingency)
        
            # Update best values if silhouette score or purity is better
            if silhouette_avg > best_silhouette_score:# whether need to >0
                best_silhouette_score = silhouette_avg
                best_eps = eps
                best_min_samples = min_samples
                best_purity = purity

# Report the optimum eps, min_samples, silhouette score, and purity
print("Optimum eps value:", best_eps)
print("Optimum min_samples value:", best_min_samples)
print("Best Silhouette Score:", best_silhouette_score)
print("Best Purity:", best_purity)
```

The result indicates that the optimum epsilon (eps) value found by DBSCAN clustering is 12, and the optimum minimum samples (min_samples) value is 16. This result is different from Q2's result which is opitmal k=2. However, the clustering solution obtained using these parameters yields a negative silhouette score of -0.2148 and a purity of 0.6082.
